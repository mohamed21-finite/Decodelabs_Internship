## Task2 Cleaning the data

## the region Column wasn't right i'm going to assign the right continent with the right country

```python
countries =df['Country'].unique().tolist()
countries
```

```python
regions = df['Region'].unique().tolist()
regions
```

```python
country_to_continent = {
    'Brazil': 'South America',
    'Germany': 'Europe',
    'South Africa': 'Africa',
    'India': 'Asia',
    'China': 'Asia',
    'Australia': 'Asia',
    'Canada': 'North America',
    'USA': 'North America',
    # ... add all countries from your list
}
```

```python
df['Region'] = df['Country'].map(country_to_continent)
```

# Task3 EDA

```python
df1 = df.copy()
```

```python
df1.describe()
```

```python
emission_by_years = df1.groupby('Year')['Co2_Emissions_MetricTons'].sum().reset_index()
maximum_5_emission_by_years = emission_by_years.nlargest(5, 'Co2_Emissions_MetricTons')
maximum_5_emission_by_years
```

```python
emission_by_sectors = df1.groupby('Industry_Type')['Co2_Emissions_MetricTons'].sum().reset_index()
maximum_5_emission_by_sectors = emission_by_sectors.nlargest(5, 'Co2_Emissions_MetricTons')
maximum_5_emission_by_sectors
```

```python
emission_by_regions = df1.groupby('Region')['Co2_Emissions_MetricTons'].sum().reset_index()
emission_by_regions.sort_values(by='Co2_Emissions_MetricTons', ascending=False).head(5)
```

# Task4 Data Visualization

```python
import matplotlib.pyplot as plt

plt.figure()

emission_by_countries = df1.groupby('Country')['Co2_Emissions_MetricTons'].sum().reset_index()

plt.bar(emission_by_countries['Country'], emission_by_countries['Co2_Emissions_MetricTons'])

plt.xlabel('Country')
plt.ylabel('CO2 Emissions (Metric Tons)')
plt.title('CO2 Emissions by Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
emission_by_sectors = df1.groupby('Industry_Type')['Co2_Emissions_MetricTons'].sum().reset_index()
plt.bar(emission_by_sectors['Industry_Type'], emission_by_sectors['Co2_Emissions_MetricTons'])
plt.xlabel('Industry Type')
plt.ylabel('CO2 Emissions (Metric Tons)')
plt.title('CO2 Emissions by Industry Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
emission_by_regions = df1.groupby('Region')['Co2_Emissions_MetricTons'].sum().reset_index()
plt.bar(emission_by_regions['Region'], emission_by_regions['Co2_Emissions_MetricTons'])
plt.xlabel('Region')
plt.ylabel('CO2 Emissions (Metric Tons)')
plt.title('CO2 Emissions by Region')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure()
emission_by_years = df1.groupby('Year')['Co2_Emissions_MetricTons'].sum().reset_index()
plt.plot(emission_by_years['Year'], emission_by_years['Co2_Emissions_MetricTons'], marker='o')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (Metric Tons)')
plt.title('CO2 Emissions by Year')
plt.tight_layout()
plt.show()
```
